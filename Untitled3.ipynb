```python
# ============================================
# Neurona / Perceptrón binario (0/1)
# z = w·x + b  y = step(z, umbral)
# ============================================

def step(z, umbral=0.0):
    return 1 if z >= umbral else 0

def neurona(x1, x2, w1, w2, b, umbral=0.0):
    z = x1*w1 + x2*w2 + b
    y = step(z, umbral)
    return z, y

    # ============================================
# Pruebas controladas
# ============================================

casos = [(0,0), (0,1), (1,0), (1,1)]

def ejecutar_prueba(nombre, w1, w2, b, umbral=0.0):
    print(f"\n=== {nombre} ===")
    print(f"w1={w1}, w2={w2}, b={b}, umbral={umbral}\n")
    for x1, x2 in casos:
        z, y = neurona(x1, x2, w1, w2, b, umbral)
        print(f"Entrada=({x1}, {x2})  z={z:.1f}  salida={y}")

# 1) AND (solo da 1 cuando (1,1))
ejecutar_prueba("PRUEBA 1: AND", w1=1, w2=1, b=-1.5, umbral=0.0)

# 2) OR (da 1 cuando al menos una entrada es 1)
ejecutar_prueba("PRUEBA 2: OR", w1=1, w2=1, b=-0.5, umbral=0.0)

# 3) Cambiar bias (mismo w, distintas b)
ejecutar_prueba("PRUEBA 3: Misma w, bias más NEGATIVO (más difícil llegar a 1)", w1=1, w2=1, b=-2.5, umbral=0.0)
ejecutar_prueba("PRUEBA 4: Misma w, bias más POSITIVO (más fácil llegar a 1)", w1=1, w2=1, b=0.5, umbral=0.0)

# 4) Cambiar pesos (mismo b, distinta w)
ejecutar_prueba("PRUEBA 5: w1 más grande (x1 influye más)", w1=2, w2=1, b=-1.5, umbral=0.0)
ejecutar_prueba("PRUEBA 6: w2 negativo (x2 resta en z)", w1=1, w2=-1, b=-0.5, umbral=0.0)

# 5) Cambiar umbral (misma w y b, pero threshold más alto)
ejecutar_prueba("PRUEBA 7: Umbral más alto (más difícil obtener 1)", w1=1, w2=1, b=-0.5, umbral=1.0)


import pandas as pd

configuraciones = [
    ("AND", 1, 1, -1.5, 0.0),
    ("OR",  1, 1, -0.5, 0.0),
    ("Bias -2.5", 1, 1, -2.5, 0.0),
    ("Bias +0.5", 1, 1, 0.5, 0.0),
    ("w1=2", 2, 1, -1.5, 0.0),
    ("w2=-1", 1, -1, -0.5, 0.0),
    ("Umbral 1.0", 1, 1, -0.5, 1.0),
]

rows = []
for nombre, w1, w2, b, umbral in configuraciones:
    for x1, x2 in casos:
        z, y = neurona(x1, x2, w1, w2, b, umbral)
        rows.append({
            "Config": nombre,
            "x1": x1,
            "x2": x2,
            "w1": w1,
            "w2": w2,
            "b": b,
            "umbral": umbral,
            "z": round(z, 2),
            "salida y": y
        })

df = pd.DataFrame(rows)
df
```
